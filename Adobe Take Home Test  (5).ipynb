Merging all datasets that have student ids for predictive modeling A

```python
from functools import reduce
student_data = [student_info, student_registration, student_assessment]
merged_student_data= reduce(lambda left,right: pd.merge(left,right,on='id_student'),student_data)
merged_student_data['final_result'].value_counts().sum()
```

Since the target variable, final_grade, is categorical, i am converting my final dataset to categorical to in order to use it for predictive modeling

```python
from sklearn import preprocessing
cat = preprocessing.LabelEncoder()
merged_student_data = merged_student_data.apply(cat.fit_transform)
```

Random Forres is a quick model to build and usually has pretty reliable results, which make Random Forest a good baseline model to chose and later to compare against. The accuracy is 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

X = merged_student_data.loc[:, merged_student_data.columns != 'final_result']
y = merged_student_data['final_result']
xTrain, xTest, yTrain, yTest = train_test_split(X, y,train_size = 0.75)

random_forrest_model = RandomForestClassifier(n_estimators=10,random_state=33)
random_forrest_model = random_forrest_model.fit(xTrain, yTrain)
train_pred = random_forrest_model.predict(xTrain)
test_pred = random_forrest_model.predict(xTest)
print("Accuracy:{0:.3f}".format(metrics.accuracy_score(yTest, test_pred)),"\n")
```

Understanding which features have the highest importance

```python
random_forrest_model.fit(X, y)
importances = random_forrest_model.feature_importances_
std = np.std([tree.feature_importances_ for tree in random_forrest_model.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the impurity-based feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
        color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```

Model Performance Evaluation

```python
from sklearn.model_selection import learning_curve, GridSearchCV

def plot_accuracy(model):

    train_sizes, train_scores, test_scores = learning_curve(model, xTrain, yTrain, n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5), verbose=0)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    
    plt.figure()
    plt.gca().invert_yaxis()
    plt.grid()
    plt.ylim(0.0, 1.1)
    plt.title("Accuracy Plot")
    plt.xlabel("Testing")
    plt.ylabel("Accuracy %")

    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1)
        
    plt.plot(train_sizes, test_mean, 'bo-', color = "r", label="Test Score")
```

```python
plot_accuracy(random_forrest_model)
```
